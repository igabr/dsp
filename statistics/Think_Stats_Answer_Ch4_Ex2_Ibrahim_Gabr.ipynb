```python
import thinkstats2
import thinkplot
import numpy as np
```

```python
random_numbers = np.random.random(1000)
```

```python
pmf = thinkstats2.Pmf(random_numbers)
thinkplot.Pmf(pmf, linewidth=0.08)
thinkplot.Show(xlabel = "Random variable", ylabel="PMF")
```

The above graph show's a uniform 'block'. I played with the line width for better visibility. This confirms that the pmf is uniform.

```python
cdf = thinkstats2.Cdf(random_numbers)
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel = "Random variable", ylabel="CDF")
```

The above shows a linear relationship for the CDF. This shows a uniform distribution
